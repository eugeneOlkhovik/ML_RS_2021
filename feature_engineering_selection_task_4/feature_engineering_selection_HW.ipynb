```python
def my_scaler(df):
    try:
        # by default std function calculates adjusted standart diviation, it means that deniminator comprise 
        # (n-1). To calculate regular standart deviation we need to set the parameter ddof to 0
        sigma = df.mean()
        delta = df.std(ddof=0)
        scaled_df = (df - sigma)/delta
        
        #StandardScaler
        scaler = StandardScaler()
        scaler.fit(df)
        skl_standart = scaler.transform(df)
        skl_standart =  pd.DataFrame(skl_standart, index=df.index, columns=df.columns)
        
        assert np.allclose(scaled_df.values, skl_standart.values, rtol=1e-6)
        print("Scaled data equals to sklearn's scaler")
        return scaled_df
        
    except NameError as er:
        print(er)
        print("StandardScaler is not imported")
            
    except (ValueError, TypeError) as error:
        print(error)
        print( "data is irrelevant")
        
    except AssertionError as err:
        print(err)
        print("Scaled data does not equal to sklearn's scaler")
    
```

```python
# fucntion to build correlation matrix
def correlation_matrix_plot(data_for_print):
    f, ax = plt.subplots(figsize=(15, 15))
    corr = data_for_print.corr()
    sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
```

```python
#function to remove highly correlated vars
def remove_highly_correlated_vars(dataset, threshold):
    col_corr = set() # Set of all the names of deleted columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j] >= threshold) and (corr_matrix.columns[j] not in col_corr):
                colname = corr_matrix.columns[i] # getting the name of column
                col_corr.add(colname)
                if colname in dataset.columns:
                    del dataset[colname] # deleting the column from the dataset
```

```python
def logistic_regression_wine(x_train, y_train, x_test):
    ud_lr = LogisticRegression(solver='liblinear', penalty='l1', random_state=1863)
    ud_fit = ud_lr.fit(x_train,y_train)
    print("Classificator labels: {0} \n Intersept Coef: {1} \n Coeff of model: {2}".format(ud_lr.classes_,
                                                                                      ud_lr.intercept_,
                                                                                     ud_lr.coef_))
    return ud_lr, ud_lr.predict(x_test)
```

```python
# отношение правильных к общему кол-ву лэйблов (accuracy_score одно и то же)
def model_accuracy(regressor, x_test, y_test):
    print('Accuracy of logistic regression classifier on test set: {:.4f}'\
          .format(regressor.score(x_test, y_test)))
    return regressor.score(x_test, y_test)
```

```python
def build_roc_curve(ud_lr, x_test,x_train, y_test,y_train):
    print("Roc-auc score: ",roc_auc_score(y_test, ud_lr.predict_proba(x_test), multi_class='ovr'))
    ud_wine_lable_score = ud_lr.fit(x_train,y_train).decision_function(x_test)
    
    classifier = OneVsRestClassifier(ud_lr)
    y_score = classifier.fit(x_train, y_train).decision_function(x_test)
# Compute ROC curve and ROC area for each class
    y = [0,1,2]
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    n_classes = 3
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test, y_score[:,i], pos_label=i)
        roc_auc[i] = auc(fpr[i], tpr[i])
    
    # First aggregate all false positive rates
    all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))

# Then interpolate all ROC curves at this points
    mean_tpr = np.zeros_like(all_fpr)
    for i in range(n_classes):
        mean_tpr += interp(all_fpr, fpr[i], tpr[i])

# Finally average it and compute AUC
    mean_tpr /= n_classes
    lw = 2
    colors = cycle(['aqua', 'darkorange', 'cornflowerblue'])
    for i, color in zip(range(n_classes), colors):
        plt.plot(fpr[i], tpr[i], color=color, lw=lw,
             label='ROC curve of class {0} (area = {1:0.2f})'
             ''.format(i, roc_auc[i]))

    plt.plot([0, 1], [0, 1], 'k--', lw=lw)
    plt.xlim([-0.01, 1.0])
    plt.ylim([-0.01, 1.05])
    
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Some extension of Receiver operating characteristic to multi-class')
    plt.legend(loc="lower right")
    plt.show()

    
```

```python
def model_estimators(x_test, y_test, pred, regressor):
    print(" Precision score: {0} \n Recall score: {1}".format(\
    round(precision_score(y_test,pred, average = 'weighted'),4),
round(recall_score(y_test,pred, average = 'weighted'),4)))
    
    print("\n Plot confusion matrix: ")
    plot_confusion_matrix(regressor, x_test, y_test)  
    
   
```
